```python
import sys
sys.path.insert(1, 'D:/kaneshi')
import kaneshi

from kaneshi.core.market_data import MarketData
from kaneshi.core.strategies import RSIOBSOneEdgeFixedStop
```

```python
m_config = {'symbol': 'BTCUSDT',
            'interval': 1,
            's_date': (2021, 1, 1),
            'e_date': (2021, 1, 5),
            'columns': 'Close',
            'price_type': 'Close',
           }

market = MarketData.from_config(**m_config)
```

```python
s_config = {
    'rsi_period': 14,
    'edge': 30,
    'stop_loss_percent': -0.01,
    'take_profit_percent': 0.01,
    'market_data': market,
}

s = RSIOBSOneEdgeFixedStop(**s_config).apply()
```

```python
s.generate_report(type_='df')
```

```python
s.plot(plot_type='plt')
```
